```python
import shutup; shutup.please()
import json
import random
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.firefox.firefox_profile import FirefoxProfile
import requests
import urllib
from tqdm import tqdm
```

```python
def input_id(id_,keys,time_before,time_after):
    global driver
    time.sleep(time_before)
    iter_error=0
    while driver.find_element(By.ID,id_).get_attribute('value')!=keys:
        try:
            driver.find_element(By.ID,id_).clear()
            time.sleep(1)
            driver.find_element(By.ID,id_).send_keys(keys)
        except:
            pass
        iter_error+=1
        if iter_error>5:
            break
    time.sleep(time_after)

def click_id_noname(id_,time_before,time_after):
    global driver
    time.sleep(time_before)
    flag=True
    iter_error = 0
    while flag:
        try:
            btn = driver.find_element(By.ID,id_)
            btn.click()
            #print(f"Клик по id {id_}")
            flag=False
        except:
            time.sleep(1)
            iter_error += 1
            if iter_error >= 5:
                try:
                    btn.click()
                    #print(f"Клик по id {id_}")
                except:
                    pass
                flag = False
        time.sleep(time_after)

def download_video(video_url):
    try:
        video_name=video_url.split("/")[-1]
        r = requests.get(video_url)
        with open(f'./vid/{video_name}', 'wb') as f:
            f.write(r.content)
        return video_name
    except:
        return ""
    
def find_end_test():
    global driver
    time.sleep(3)
    trigger=["Вы не сдали","Вы сдали"]
    h3=driver.find_elements(By.TAG_NAME,"h3")
    for h in h3:
        if h.text=="Тест не пройден":
            try:
                btn=driver.find_elements(By.TAG_NAME,"button")
                for b in btn:
                    if b.text=="ЗАВЕРШИТЬ ТЕСТ":
                        b.click()
            except:
                pass
            try:
                btn=driver.find_elements(By.TAG_NAME,"button")
                for b in btn:
                    if b.text=="ЗАВЕРШИТЬ":
                        b.click()
            except:
                pass
            driver.get("https://pddtest.kz/tests")
            click_id_noname(":R1rlalal4m:",2,2)
        elif h.text in trigger:
            driver.get("https://pddtest.kz/tests")
            click_id_noname(":R1rlalal4m:",2,2)         
```

```python
iteration=input("Введите количество циклов:")
            
options=Options()
firefox_profile = FirefoxProfile()
options.profile = firefox_profile
driver = webdriver.Firefox(options=options)

url="https://pddtest.kz/auth"
driver.get(url=url)

input_id(":R6mkdal4m:","",5,2)
click_id_noname(":Remkdal4m:",2,2)
input_id(":R1lamkdal4m:","",2,2)
click_id_noname(":Remkdal4m:",2,2)
click_id_noname(":r2:",2,2)

with open('otvet.json') as f:
    file_content = f.read()
    otvet = json.loads(file_content)
```

```python
for n in tqdm(range(int(iteration))):
    find_end_test()
    lang=driver.find_elements(By.CLASS_NAME,"css-hkidb")
    for l in lang:
        if l.text=="KZ":
            l.click()
    question=driver.find_elements(By.CLASS_NAME,"question")[0]
    h3=""
    h3=question.find_elements(By.CLASS_NAME,"MuiTypography-h3")[0].text
    video_url=""
    video_url=question.find_elements(By.TAG_NAME,"source")[0].get_attribute("src")
    video_name=download_video(video_url)

    answers_btn=driver.find_elements(By.CLASS_NAME,"answers")[0].find_elements(By.CLASS_NAME,"text")
    answers=[]
    for i in range(len(answers_btn)):
        answers.append(answers_btn[i].text)

    sovp=-1
    for i in range(len(otvet)):
        if otvet[i]["h3"]==h3 and \
        otvet[i]["video_name"]==video_name and \
        len(otvet[i]["ansvers"])==len(answers):
            kolvo_sovp_answers=0
            for ans in answers:
                if ans in otvet[i]["ansvers"]:
                    kolvo_sovp_answers+=1
            if kolvo_sovp_answers==len(answers):
                sovp=i
                break

    if sovp==-1:
        ans_i_rand=0
        ans_i_rand=random.randint(0,len(answers_btn)-1)
        answers_btn[ans_i_rand].click()
        time.sleep(3)
        question=driver.find_elements(By.CLASS_NAME,"question")[0]
        video_url2=""
        video_url2=question.find_elements(By.TAG_NAME,"source")[0].get_attribute("src")
        video_name2=download_video(video_url2)
        true_answer=driver.find_elements(By.CLASS_NAME,"answers")[0].find_elements(By.CLASS_NAME,"css-1qboli5")[0].\
        find_elements(By.CLASS_NAME,"text")[0].text
        explanation=driver.find_elements(By.CLASS_NAME,"explanation")[0]
        explanation_text=explanation.find_elements(By.TAG_NAME,"p")[0].text
        time.sleep(5)
        explanation.find_elements(By.TAG_NAME,"button")[0].click()
        slovar={}
        slovar["h3"]=h3
        slovar["video_name"]=video_name
        slovar["video_name2"]=video_name2
        slovar["ansvers"]=answers
        slovar["true_answer"]=true_answer
        slovar["explanation_text"]=explanation_text
        otvet.append(slovar)
        with open('otvet.json', 'w', encoding='utf8') as f:
            f.write(json.dumps(otvet,ensure_ascii=False))
    else:
        for a in range(len(answers)):
            if otvet[sovp]["true_answer"]==answers[a]:
                answers_btn[a].click()
        time.sleep(5)
        explanation=driver.find_elements(By.CLASS_NAME,"explanation")[0]
        explanation.find_elements(By.TAG_NAME,"button")[0].click()
```
